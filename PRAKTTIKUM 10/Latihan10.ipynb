Selain itu, set juga **menghapus elemen duplikat secara otomatis**, sehingga “Alpin” hanya ditampilkan satu kali meskipun ditulis dua kali.

```python
mahasiswa = ["Saci", "Salma", "Pipi", "Zalda", "Alpin"]
mahasiswa.append("Darmawati")
mahasiswa.insert(3, "Adee")
mahasiswa.pop()
mahasiswa.sort(reverse=True)
mahasiswa.remove("Alpin")
for nama in mahasiswa:
    print("Nama Mahasiswa: ", nama)
    
print(mahasiswa)
```

#### Analisis

Setelah itu, program menampilkan setiap nama mahasiswa satu per satu serta mencetak isi list terakhir.


```python
mahasiswa = ["Saci", "Salma", "Pipi", "Zalda", "Alpin"]
mahasiswa.append("Darmawati")
mahasiswa.insert(3, "Adee")
mahasiswa.pop()
mahasiswa.sort(reverse=True)
mahasiswa.remove("Alpin")

i = 1
for nama in mahasiswa:
    print("Nama Mahasiswa ke-", i ,": ", nama)
    i+=1
    
print(mahasiswa)
```

#### Analisis

```python
person = {"name": "SalmaPutri", "hp": "080808", "kelas": "TEKOM F"}
print(person)
```

#### Analisis
Kode tersebut membuat **dictionary** bernama **person** yang menyimpan tiga pasangan **key-value**, yaitu `"name": "SalmaPutri"`, `"hp": "080808"`, dan `"kelas": "TEKOM F"`. Perintah `print(person)` akan menampilkan seluruh isi dictionary tersebut di layar.

```python
person = {"name": "SalmaPutri", "hp": "080808", "kelas": "TEKOM F"}
for key, value in person.items():
    print(key, ":", value)
```

#### Analisis
Kode tersebut membuat **dictionary** bernama **person** yang berisi tiga pasangan key-value, yaitu `"name": "SalmaPutri"`, `"hp": "080808"`, dan `"kelas": "TEKOM F"`. Selanjutnya, perulangan `for key, value in person.items():` digunakan untuk mengambil setiap pasangan key dan nilainya dari dictionary. Metode `items()` mengembalikan elemen dalam bentuk tuple `(key, value)`, sehingga keduanya dapat dipisahkan dan ditampilkan dengan format **key : value**. Hasilnya, program mencetak setiap data dalam dictionary pada baris yang berbeda, menampilkan nama, nomor hp, dan kelas secara terpisah.

```python
i = 1
while i <= 9:
    print(i)
    i += 2
```

#### Analisis
Kode tersebut memakai perulangan while untuk menampilkan angka dari 1 hingga 9 dengan kenaikan 2 setiap kali perulangan dijalankan. Variabel i diawali dengan nilai 1, kemudian nilainya bertambah 2 di setiap iterasi, sehingga angka yang dicetak berturut-turut adalah 1, 3, 5, 7, dan 9. Proses perulangan berhenti ketika nilai i sudah lebih besar dari 9.


```python
i = 1
while i <= 13:
    if i == 7:
        print("WOW")
    else:
        print(i)
    i +=2
```

#### Analisis
Kode tersebut menjalankan perulangan while dengan kenaikan nilai sebesar 2 di setiap iterasi, dimulai dari 1 hingga 13. Ketika nilai i bernilai 7, program menampilkan teks "WOW", sedangkan pada nilai lainnya program mencetak angka seperti biasa. Hasil akhirnya menampilkan urutan output 1, 3, 5, WOW, 9, 11, dan 13.


```python
for i in range(1, 11):
    if i <= 8:
        if i % 2 == 1:
            print("hidup")
        else:
            print("mahasiswa")
    else:
        print(i)
```

#### Analisis
Kode tersebut menjalankan perulangan for dari angka 1 sampai 10. Di dalamnya terdapat percabangan bersarang yang memeriksa nilai i. Jika i kurang dari atau sama dengan 8, program akan mengecek apakah i termasuk bilangan ganjil atau genap. Bila i ganjil (i % 2 == 1), program menampilkan kata "hidup", sedangkan jika genap, menampilkan kata "mahasiswa". Untuk nilai i yang lebih besar dari 8, yaitu 9 dan 10, program langsung mencetak angkanya. Hasil akhirnya berupa kombinasi antara kata dan angka sesuai dengan kondisi yang terpenuhi.


```python
for i in range(1, 14):
    print(i)
    if i % 3 == 2:  
        print("Ha")
```

#### Analisis
Kode tersebut menjalankan perulangan for dari angka 1 hingga 13. Setiap nilai i akan dicetak terlebih dahulu, kemudian program memeriksa kondisi apakah i memiliki sisa bagi 2 ketika dibagi 3 (i % 3 == 2). Jika kondisi tersebut benar, maka program menampilkan kata "Ha" setelah angka tersebut. Dengan demikian, setiap kali nilai i menghasilkan sisa 2 saat dibagi 3, akan muncul tambahan output "Ha". Hasil akhirnya berupa deretan angka yang disertai kata "Ha" pada posisi tertentu sesuai kondisi yang terpenuhi.


```python
for i in range(1, 14):
    if i not in [3, 5, 6, 9, 10, 12]:
        print(i)
        
    if i in [2, 5, 8, 11]:
        print("Ha")

    if i in [4, 9]:
        print("Hi")
```

#### Analisis
Kode tersebut menjalankan perulangan dari angka 1 hingga 13 dan menampilkan output berupa angka atau kata sesuai dengan kondisi yang ditetapkan. Jika nilai i tidak berada dalam daftar [3, 5, 6, 9, 10, 12], maka program mencetak angkanya. Untuk nilai i yang termasuk dalam [2, 5, 8, 11], program menampilkan kata "Ha", sedangkan untuk nilai i dalam [4, 9], program mencetak "Hi". Dengan demikian, hasil akhir yang dihasilkan berupa kombinasi antara angka dan teks sesuai dengan aturan yang telah ditentukan dalam program.
