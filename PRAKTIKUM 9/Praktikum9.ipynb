## Global Variables

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

### Analisis
Kode tersebut mendefinisikan variabel `x` berisi string **“awesome”** di luar fungsi, sehingga menjadi variabel global. Fungsi `myfunc()` mencetak kalimat dengan menggabungkan teks **“Python is ”** dan nilai `x`. Saat `myfunc()` dipanggil, output yang muncul adalah **“Python is awesome”**.


```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

### Analisis
Kode tersebut memperlihatkan perbedaan **variabel lokal** dan **global**.
Di dalam fungsi, `x` diberi nilai **“fantastic”** sehingga saat `myfunc()` dipanggil, outputnya **“Python is fantastic”**.
Namun, variabel `x` di luar fungsi tetap **“awesome”**, sehingga baris terakhir mencetak **“Python is awesome”**.


## The global Keyword

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis
Program tersebut mendefinisikan fungsi `myfunc()` yang menggunakan kata kunci `global` agar variabel `x` dapat diakses di luar fungsi. Di dalam fungsi, `x` diberi nilai `"fantastic"`, dan setelah fungsi dipanggil, nilai tersebut digunakan dalam perintah `print("Python is " + x)`, sehingga menghasilkan output **"Python is fantastic"**.


```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis
Program tersebut menetapkan `x = "awesome"`, lalu fungsi `myfunc()` mengubah nilai global `x` menjadi `"fantastic"`. Setelah fungsi dipanggil, perintah `print("Python is " + x)` menampilkan output **"Python is fantastic"**.

## Type Conversion

```python
x = 1    # int
y = 2.8  # float
z = 1j   # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

### Analisis
Program tersebut menunjukkan konversi tipe data di Python. Variabel `x` yang bertipe **int** diubah menjadi **float** pada `a = float(x)`, menghasilkan `1.0`. Variabel `y` yang bertipe **float** diubah menjadi **int** pada `b = int(y)`, menghasilkan `2` (nilai desimal dibuang). Variabel `x` juga diubah menjadi **complex** pada `c = complex(x)`, menghasilkan `(1+0j)`. Perintah `print(type(...))` menegaskan tipe baru: `a` bertipe **float**, `b` bertipe **int**, dan `c` bertipe **complex**.

## The Walrus Operator

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

### Analisis
Program tersebut menunjukkan dua cara untuk menghitung dan memeriksa jumlah elemen dalam list `numbers = [1, 2, 3, 4, 5]`. Pada blok pertama, `len(numbers)` disimpan di variabel `count` lalu dibandingkan dengan 3; karena jumlah elemen **5 > 3**, dicetak **"List has 5 elements"**. Pada blok kedua, operator **walrus `:=`** digunakan untuk menghitung panjang list langsung dalam kondisi `if`; hasilnya sama, dicetak **"List has 5 elements"**, tetapi lebih ringkas.

## The Python Match Statement

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

### Analisis 
Program ini menggunakan **struktur match-case** untuk menentukan hari dalam seminggu berdasarkan nilai `day = 4`. Karena nilai `day` sama dengan 4, program mengeksekusi blok `case 4` dan menampilkan pesan **“Thursday”**. Struktur ini mirip dengan penggunaan `if-elif-else` tetapi lebih rapi untuk banyak kondisi diskrit.


## Default Value

```python
day = 4
match day:
  case 6:
    print("Today is Saturday")
  case 7:
    print("Today is Sunday")
  case _:
    print("Looking forward to the Weekend")
```

### Analisis
Program ini menggunakan **struktur match-case** untuk mengevaluasi nilai `day = 4`. Karena 4 tidak sesuai dengan `case 6` atau `case 7`, program mengeksekusi `case _`, yang berfungsi sebagai **default**, sehingga menampilkan pesan **“Looking forward to the Weekend”**.


## Combine Values

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

### Analisis
Program ini menggunakan **struktur match-case** dengan operator **or (`|`)** untuk mengelompokkan beberapa nilai. Karena `day = 4` termasuk dalam `case 1 | 2 | 3 | 4 | 5`, program mengeksekusi blok ini dan menampilkan pesan **“Today is a weekday”**, menandakan bahwa hari tersebut adalah hari kerja.


## If Statements as Guards

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

### Analisis
Program ini menggunakan **match-case** dengan **guard condition** (`if`) untuk memeriksa kombinasi hari dan bulan. Variabel `day = 4` dan `month = 5`, sehingga cocok dengan `case 1 | 2 | 3 | 4 | 5 if month == 5`. Akibatnya, program mengeksekusi blok ini dan menampilkan pesan **“A weekday in May”**.
